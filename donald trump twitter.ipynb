
## Question 6 -Find % of accounts with more than 50% positive tweets about Donald Trum

```python
donald_trump_tweet_user_counts = df_donald_trump_tweets.user_id.value_counts()#value counts of each user id of donlald trump's tweets data
donald_trump_tweet_user_counts.head()
```

```python
donald_trump_positive_tweet_user_counts= df_donald_trump_positive_tweets.user_id.value_counts()#value counts of each user id of donald trump's positive tweet data
donald_trump_positive_tweet_user_counts.head()
```

```python
#Find % of accounts with more than 50% positive tweets about Donald Trump.
array = []
for key , value in donald_trump_positive_tweet_user_counts.items():

    #print(key,donald_trump_positive_tweet_user_counts[key],donald_trump_tweet_user_counts[key])
    var = value/donald_trump_tweet_user_counts[key]*100
    #print(var , "Percentage")
    if var >=50:
        array.append(key)
print("Following are the users with more than 50 % positive tweets about Donald Trump")        
array                
```

## Question7-Create distribution of all donald trump tweets for each hour of day, and compare standard deviations

```python
#resample tweets by minutes and then calculate std hourly
tweets_by_minute= df_donald_trump_tweets.resample('T' , how = {"text" : "count"})
b = tweets_by_minute.resample("H").std()
b
```

```python
b.plot(figsize=(15,10),color = "green")
plt.xlabel("Time" , fontsize=20)
plt.ylabel("Standard deviation" , fontsize=20)

plt.title('Hourly standard deviation plot' , fontsize=20)
```

## Question 8-Overlay hour of day distributions with accounts with more than 50% positive tweets. See if you can use this overlay to identify timezones of USA who are more supportive of Donald Trump.

```python

P = df_donald_trump_positive_tweets[df_donald_trump_positive_tweets["user_id"].isin(["23022687", "29442313", "1058717720", "343358379", "818975124460335104"])]
Hourly_tweets_positive= P.id.resample('H').count()

#resample tweets by minutes and then calculate std hourly
Total_Hourly_tweets= df_donald_trump_tweets.id.resample('H').count()
```
